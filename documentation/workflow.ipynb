# How to create an experiment and visualize it? 

## 1. The config creation

Create a config and specify what dynamics, and parameter ranges you want to test.

## 2. Run the experiment

We will demonstrate to run a small experiment from the `config/default.yaml`config file

```python
from cbx_experiments.runner import Runner
from cbx_experiments.experiment_generation import create_experiment_config

experiment_config = create_experiment_config("../configs/default.yaml")
runner = Runner(experiment_config, result_dir="../results")
experiment_result = runner.run_experiment()
```

## 3. Visualize the result

Here we have a lot of possibilities. First, we load the dataframe from the .csv result.

```python
from cbx_experiments.experiment_result import ExperimentResult

experiment_result = ExperimentResult.from_csv("../results/default.csv")
df = experiment_result.to_dataframe()
plot_dir = "../results/plots"
```

1. We visualize the sigma with the maximum success probability for different lambdas.

```python
from cbx_experiments.visualization import plot_sigma_max_to_lambda
plot_sigma_max_to_lambda(df, plot_dir)
```

2. We can plot the average success probability for different methods for varying number of particles.

```python
from cbx_experiments.visualization import plot_N

plot_N(df)
```

3. We can plot the success probability for different dynamics for varying sigma values.

```python
from cbx_experiments.visualization import plot_sigma

plot_sigma(df)
```

```python
from cbx_experiments.visualization import plot_2d_sigma_lamda

plot_2d_sigma_lamda(df)
```
